NEW FUNCTION: ADD_PRODUCT

```python
def add_product(name, price, quantity):
    """Adds a new product or updates the quantity of an existing one."""
    #¿IT ALREADY EXIST?
    for material in inventory:
        if material['name'] == name.lower():
            material['quantity'] += quantity
            return f"Stock of {name} updated. Total quantity: {material['quantity']}" 
    
    #NEW MATERIAL FOUNDED.
    new_material = {
        'name' : name.lower(),
        'price' : price,
        'quantity' : quantity
    }

    inventory.append(new_material)
    return f"New material {name} added."
```

NEW FUNCTION: SHOW_INVENTORY

```python
def show_inventory():
    """Displays all products currently in the inventory."""
    for material in inventory:
        print (f"Nombre: {material['name']}, Precio: {material['price']}, Cantidad: {material['quantity']}.")
```

NEW FUNCTION: SEARCH_PRODUCT

```python
def search_product(name):
    """Searches for a product by name and returns its details."""
    for material in inventory:
        if material['name'] == name.lower():
            print (f"Nombre: {material['name']}, Precio: {material['price']}, Cantidad: {material['quantity']}.")
            return
    print (f"Producto no figura en stock")
    
```

NEW FUNCTION: UPDATE INVENTORY

```python
def update_inventory(name, quantity):
    """Updates the stock level of a specific product."""
    for material in inventory:
        if material['name'] == name.lower():
            #Calculate new stock:
            new_stock = material['quantity'] + quantity
            if new_stock < 0:
                print (f"Stock de {name} insuficiente. Stock actual: {material['quantity']} unidades") 
                return
            else:
                material['quantity'] += quantity
            print (f"El producto {material['name']} se ha actualizado a {material['quantity']}") 
            return
    print (f"Producto no figura en stock") 
```

```python
admin_name = input("Indica el producto a actualizar").strip().lower()
admin_quantity = int(input ("Introduzca cantidad. Use (+) o (-) para indicar si añade o quita"))

update_inventory(admin_name, admin_quantity)
```

NEW FUNCTION: REMOVE PRODUCT

```python
def remove_product(name):
    """Removes a product entirely from the inventory."""
    for material in inventory:
        if material['name'] == name.lower():
            inventory.remove(material)
            return f"Success. Material {name} removed."
    return "Material not found in the inventory"
```

NEW FUNCTION: CALCULATE TOTAL INVENTORY VALUE

```python
def calculate_total_inventory_value():
    """Calculates the total monetary value of all items in stock."""
    total_value = 0.00  ## TENGO TOTAL VALUES AL PRINCIPIO, PREGUNTAR A YANELIS
    for material in inventory:
        total_value = total_value + (material['quantity'] * material ['price'])
    return total_value
```

NEW FUNCTION: PROCESS PURCHASE

```python

def process_purchase():
    """Allows a customer to select products, updates inventory, and calculates the total cost."""
    cart = []
    cart_total = 0.00
    show_inventory()
    #Shopping switch on
    shopping = True 
    while shopping:
        user_material = input ("¿Qué material desea comprar?. Si desea terminar escriba <salir>").lower().strip()
        if user_material == "salir": #Client can end loop
            shopping = False
        
        else:
            encontrado = False #switch "looking for"
            #Looking for the material:
            for material in inventory:
                if user_material == material['name'].lower():
                    encontrado = True #We have found the material
                    try:
                        user_quantity = int(input ("¿Cuánta cantidad desea?"))
                        #Calculate stock:
                        new_stock = material['quantity'] - user_quantity
                        print (f"Se ha seleccionado {user_quantity} unidad(es) de {user_material}.")
                        if new_stock < 0:
                            print (f"Stock de {user_material} insuficiente. Stock actual: {material['quantity']} unidad(es)") 
                        else:
                            #Using previous function to update inventory. Always negative (client is reducing material)
                            update_inventory(user_material, user_quantity * -1)
                    
                            material_selected = {
                                'name': material['name'],
                                'price': material['price'],
                                'quantity': user_quantity
                            }
                            #Add material to cart:
                            cart.append(material_selected)
                            print (f"Se ha/n añadido al carrito {user_quantity} unidad(es) de {user_material}.")    
                        break
                    except:
                        print("Error: Por favor, introduce un número entero (ej: 5).")
            if encontrado == False: #Material don't correspond with the list.
                print(f"El producto '{user_material}' no existe en el inventario. Comprueba el inventario e inténtalo de nuevo.")
                input("--> Material no encontrado. Pulsa ENTER para elegir otro...")

    #Calculate total cost and print ticket
    print("\n------ RESUMEN DE TU COMPRA ------")
    for material in cart:
        subtotal = (material['quantity'] * material ['price'])
        cart_total = cart_total + subtotal
        print(f"- {material['name']}: {material['quantity']} unidad(es)   | {subtotal}€")

    print("-----------------------------------")
    print(f"TOTAL A PAGAR: {cart_total}€")
    print("¡Gracias por elegirnos!\n")

    return cart_total, cart 
  
```

```python
#PRUEBA PROCESS PURCHASE 
#WE ASING VALUES FOR FURTHER USES
cart_total, cart = process_purchase()  
```

```python
def process_payment():
    paid = False #switch for payment
    while not paid:
        try: #"try" will encapsulate errors considering a bad input 
            customer_payment =  float(input ("¿Cuánto dinero entrega"))

            if customer_payment < cart_total:
                print (f"Error. Montante insuficiente, el valor de la compra es de {cart_total}")
        
            else:
                change = customer_payment - cart_total
                print("\n------ RESUMEN DE TU COMPRA ------")  #PRINT TICKET
                for material in cart:
                    subtotal = (material['quantity'] * material ['price'])
                    print(f"- {material['name']}: {material['quantity']} unidad(es)   | {subtotal}€")
                print("-----------------------------------")
                print(f"TOTAL A PAGAR: {cart_total}€")
                print (f"Cliente entrega {customer_payment} €")
                print (f"Compra completada con éxito. Tu vuelta es {change} €")
                print("¡Gracias por su compra!\n")
                paid = True #Payment succeed

        except:
            print("Error. Por favor, introduce un número (ej: 5.00).") #Value is str
```

NEW FUNCTION: ADD CUSTOMER

```python
def add_customer(name, email): #We have to add name and email considering the original dictionary
    """Adds a new customer to the original list."""
    
    #WE ARE USING NAMES AS KEYS.
    #WE ADD A STEP TO PROTECT THE DATA THAT ALREADY EXIST. 

    new_name = name.title().strip()
    if new_name in customers:
        return f"Error: El cliente {new_name} ya existe en el sistema."
    else:
        customers[name.title().strip()] = {
            'email' : email.lower().strip(),
            'purchases' : []
        }

        return f"New customer {name} added."
```
